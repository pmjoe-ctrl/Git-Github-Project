1. What is File Handling?
File handling means working with files in your program. This includes:

Opening files
Reading data from files
Writing data to files
Closing files
Python makes it easy to work with files using built-in functions and libraries.

2. Opening a File
To work with a file, you first need to open it using Python’s open() function. Here’s the basic syntax:

file = open("filename.txt", "mode")
filename.txt: The name of the file you want to open (or its path).
mode: Tells Python what you want to do with the file (read, write, etc.).
Common File Modes: Here are the most commonly used modes:
Mode	Description
'r'	Read mode: Opens the file for reading (default).
'w'	Write mode: Opens the file for writing. Creates a new file if it doesn’t exist or overwrites the existing file.
'a'	Append mode: Opens the file for appending. Adds new data to the end of the file.
'x'	Exclusive creation mode: Creates a new file. Fails if the file already exists.
'b'	Binary mode: Used for non-text files like images (e.g., 'rb' or 'wb').
3. Reading from a File
Once you open a file in read mode ('r'), you can read its contents. Here’s how:

# Open the file in read mode
with open("example.txt", "r") as file:
    content = file.read()  # Reads the entire file
    print(content)
The with statement automatically closes the file after reading, so you don’t have to worry about forgetting to close it.
file.read() reads the entire file as a single string.
Reading Line by Line
If the file is large, you can read it line by line:

with open("example.txt", "r") as file:
    for line in file:
        print(line)  # Prints each line one by one
4. Writing to a File
To write data to a file, open it in write mode ('w') or append mode ('a').

4.1 Write Mode ('w')
This mode overwrites the file if it already exists. If the file doesn’t exist, it creates a new one.

with open("example.txt", "w") as file:
    file.write("Hello, World!")  # Writes this text to the file
4.2 Append Mode ('a')
This mode adds new data to the end of the file without deleting the existing content.

with open("example.txt", "a") as file:
    file.write("\nThis is a new line.")  # Adds this text to the end of the file
5. Working with Different File Types
5.1 CSV Files
CSV (Comma-Separated Values) files are commonly used in data science to store tabular data. Python’s numpy library provides a function called genfromtxt() to load data from CSV files.

What is genfromtxt()?
genfromtxt() is a function in the numpy library that reads data from a text file (like a CSV) and converts it into a NumPy array. It’s very useful for working with numerical data.

Reading a CSV File with genfromtxt()
Here’s how you can use genfromtxt() to read a CSV file:

import numpy as np

# Load data from a CSV file
data = np.genfromtxt("data.csv", delimiter=",")

# Print the data
print(data)
data.csv: The name of the CSV file.
delimiter=",": Specifies that the values in the file are separated by commas.
The data is loaded into a NumPy array, which you can use for calculations or analysis.
6. Working with JSON Files
JSON (JavaScript Object Notation) is a popular format for storing and exchanging data. Python’s json module makes it easy to work with JSON files.

Reading a JSON File
import json
with open("data.json", "r") as file:
    data = json.load(file)  # Loads JSON data into a Python dictionary
    print(data)
Writing to a JSON File
data = {"name": "John", "age": 30, "city": "New York"}
with open("data.json", "w") as file:
    json.dump(data, file)  # Writes the dictionary to a JSON file
7. Closing Files
When you’re done working with a file, you should always close it to free up system resources. If you use the with statement, Python will automatically close the file for you. For example:

with open("example.txt", "r") as file:
    content = file.read()
    print(content)
# The file is automatically closed here
If you don’t use with, you must close the file manually:

file = open("example.txt", "r")
content = file.read()
print(content)
file.close()  # Don't forget to close the file!

8. Deleting Files
To delete a file, use Python’s os module:

import os
os.remove("example.txt")  # Deletes the file
